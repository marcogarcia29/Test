```python
import pandas as pd 

math_student = pd.read_csv("./data/student-mat.csv", sep=";")
por_student = pd.read_csv("./data/student-por.csv", sep=";")
```

```python
merged = math_student.merge(por_student, how='outer')
merged
```

# Quantidade de alunos da escola GP 

```python
school_gp = merged.loc[
    (merged.school == 'GP')
]
school_gp
```

# Quantidade de alunos da escola GP que receberam ajuda extra-curricular

```python
school_gp_helped = merged.loc[
    (merged.school == 'GP') &
    (merged.schoolsup == 'yes')
]
school_gp_helped
```

# % de Alunos do colegio GP com ajuda extra-curricular

```python
percentage = school_gp_helped['school'].count() / school_gp['school'].count()
percentage
```
