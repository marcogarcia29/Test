```python
import pandas as pd 

math_student = pd.read_csv("./data/student-mat.csv", sep=";")
por_student = pd.read_csv("./data/student-por.csv", sep=";")
```

```python
merged = math_student.merge(por_student, how='outer')
merged
```

# Mostrar apenas alunos com notas baixas e que tem mãe professora

```python
low_grades = merged.loc[
    (merged.Mjob == 'teacher') &
    (merged.G1 < 6) &
    (merged.G2 < 6)
]
```
